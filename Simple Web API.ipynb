# A very, very brief introduction to Web APIs

## What's an API?

API is an acronym for:

Application  
Program  
Interface  

In much the same way as the user interface allows you to interact with a program (example: web browser window, email, or Excel), an API allows one program to interact with another program.

A web API is a special type of API that allows 2 two computers on the internet to interact with each other.


There are web APIs to do nearly everything you could imagine online.
